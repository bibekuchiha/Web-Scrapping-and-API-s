```python
from bs4 import BeautifulSoup as soup
import requests
```

```python
url = 'https://www.passiton.com/inspirational-quotes?page=2'
```

```python
response = requests.get(url)
```

```python
print(response.content)
```

```python
#html parser
soup = soup(response.content,'html.parser')
```

```python
print(soup)
```

```python
div = soup.findAll("div", {"class": "col-6"})
```

```python
print(div)
```

```python
div1 = div[0]
```

```python
print(div1)
```

```python
div1.img.attrs['src']
```

```python
with open('inspiration1.jpg','wb') as f:
    img_url = div1.img.attrs['src']
    response = requests.get(img_url)
    
    f.write(response.content)
```

```python
for i, div1 in enumerate(div):
    with open('img/inspiration{}.jpg'.format(i),'wb') as f:
        img_url = div1.img.attrs['src']
        response = requests.get(img_url)

        f.write(response.content)
```
